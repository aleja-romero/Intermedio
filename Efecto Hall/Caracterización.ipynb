```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
I = [0, 0.05, 0.1, 0.15, 0.17, 0.2, 0.25, 0.28, 0.3, 0.35, 0.37]  
B = [0, 27, 50.1, 75.8, 88.4, 102, 130.8, 147.8, 162.1, 186.4, 191.9]  
```

```python
slope, intercept, r_value, p_value, std_err = linregress(I, B)

slope_uncertainty = std_err
intercept_uncertainty = std_err * np.sqrt(np.sum(np.square(I)) / len(I))

fig, ax1 = plt.subplots(figsize=(10, 6))


ax1.plot(I, B, 'o', label="Datos experimentales", color='blue')


B_fit = [slope * i + intercept for i in I]
ax1.plot(I, B_fit, '-', label=f"Regresión lineal: B = {slope:.2f} ± {slope_uncertainty:.2f} * I + {intercept:.2f} ± {intercept_uncertainty:.2f}", color='red')


ax1.set_xlabel("I (A)")
ax1.set_ylabel("B (mT)")
ax1.set_title("Regresión lineal")
ax1.legend()


plt.grid(True)
plt.show()

```

```python
Residuales = []
for i in range(len(I)):
    y_feet = 530.51*I[i] -1.40
    y = B[i]
    Residuales.append(y_feet-y)


fig, ax1 = plt.subplots(figsize=(10, 6))


ax1.plot(I, Residuales, 'o', label="Residuales", color='blue')

ax1.set_xlabel("I (A)")
ax1.set_ylabel("Residuales (mT)")
ax1.set_title("Residuales")
ax1.legend()


plt.grid(True)
plt.show()
```
