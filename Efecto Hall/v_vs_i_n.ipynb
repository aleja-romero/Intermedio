```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
x = [-30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30]

y_1 = [50, 42.6, 37.7, 30.9, 23.7, 15.4, 10.9, 3.2, -3.9, -13, -17.5, -25.7, -30.5]
y_2 = [47.6, 42, 35.1, 28.1, 21.8, 16.4, 10.4, 1.1, -3.2, -11.3, -14.2, -22.7, -30.5]
y_3 = [45.6, 38.6, 34.8, 28.2, 21.5, 15.8, 10.1, 3.9, -2.7, -9.3, -13.3, -21.1, -27.9]
y_4 = [43, 36.7, 30.7, 26.7, 20.3, 13.6, 7.6, 2.4, -3.9, -7.6, -13.9, -18.5, -23.8]
y_5 = [40.7, 36.3, 29.3, 24.3, 19.4, 13.9, 7.5, 2.8, -1.8, -8.1, -11.5, -17.7, -21.9]


 
labels = ["B=230", "B=215", "B=200", "B=185", "B=170"]
y_values = [y_1, y_2, y_3, y_4, y_5]

#
plt.figure(figsize=(10, 6))

for i, y in enumerate(y_values):

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    
    y_fit = [slope * xi + intercept for xi in x]
    
    
    plt.plot(x, y, 'o', label=f"{labels[i]}")  
    plt.plot(x, y_fit, '-', label=f"{labels[i]} (Regresión)")
    
    
    plt.text(x[-1] + 2, y_fit[-1], 
             f"y = {slope:.2f} ± {std_err:.2f}x + {intercept:.2f} ± {std_err:.2f}",
             color=plt.gca().lines[-1].get_color(),
             verticalalignment='bottom', horizontalalignment='left', fontsize=10)


plt.xlabel("I_p (mA)")
plt.ylabel("V_h (mV)")
plt.title("V_h vs I_p con B=cte para placa tipo n")
plt.legend()
plt.grid(True)


plt.show()
```

```python
Residuales = []
for i in range(len(x)):
    y_feet = -1.37*x[i] +9.52
    y = y_1[i]
    Residuales.append(y_feet-y)


fig, ax1 = plt.subplots(figsize=(10, 6))


ax1.plot(x, Residuales, 'o', label="Residuales", color='blue')

ax1.set_xlabel("I_p (mA)")
ax1.set_ylabel("Residuales (mV)")
ax1.set_title("Residuales B=230")
ax1.legend()


plt.grid(True)
plt.show()
```
