```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
x = [-300, -250, -200, -150, -100, -50, 0, 50, 100, 150, 200, 250, 300]

y_1 = [-52.4, -45.2, -39.1, -31.5, -23.5, -15.3, -7.5, 15.6, 22.7, 30.6, 38.0, 45.4, 53.4]
y_2 = [-37.6, -33.3, -28.4, -23.4, -17.8, -13.0, -7.6, 12.9, 17.5, 23.5, 28.3, 32.8, 38.2]
y_4 = [18.5, 14.1, 10.3, 6.1, 3.6, 0.6, -0.5, -3.2, -5.3, -7.9, -9.8, -13.8, -18.3]
y_5 = [36.4, 28.7, 22.1, 15.1, 6.6, -1.4, -0.8, -1.3, -7.2, -13.7, -21.6, -28.0, -36.4]


 
labels = ["I_p=-25", "I_p=-15", "I_p=15", "I_p=25"]
y_values = [y_1, y_2, y_4, y_5]

#
plt.figure(figsize=(10, 6))

for i, y in enumerate(y_values):

    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    
    y_fit = [slope * xi + intercept for xi in x]
    
    
    plt.plot(x, y, 'o', label=f"{labels[i]}")  
    plt.plot(x, y_fit, '-', label=f"{labels[i]} (Regresión)")
    
    
    plt.text(x[-1] + 2, y_fit[-1], 
             f"y = {slope:.2f} ± {std_err:.2f}x + {intercept:.2f} ± {std_err:.2f}",
             color=plt.gca().lines[-1].get_color(),
             verticalalignment='bottom', horizontalalignment='left', fontsize=10)


plt.xlabel("B (mT)")
plt.ylabel("V_h (mV)")
plt.title("V_h vs B con I_p=cte para placa tipo n")
plt.legend()
plt.grid(True)


plt.show()
```
