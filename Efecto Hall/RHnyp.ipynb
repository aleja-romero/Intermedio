```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
x = [170,185,200,215,230]

y_1 = [-6.18, -6.05, -6.05, -6.0, -5.96]
y_2 = [2.47, 2.49, 2.35, 2.47, 2.35]
y_3 = -6.05
y_4 = 2.43

y_values = [y_1, y_2]

#
plt.figure(figsize=(10, 6))

for i, y in enumerate(y_values):
    
    plt.plot(x, y, 'o')  

plt.axhline(y=y_3, color='red', linestyle='--', label='Rh promedio para Ge-n = -6.05')
plt.axhline(y=y_4, color='red', linestyle='--', label='Rh promedio para Ge-p = 2.43')


plt.xlabel("B (mT)")
plt.ylabel(r"$R_H \, (\times 10^{-3})$")
plt.title("R_H en función del campo magnético")
plt.legend()
plt.grid(True)


plt.show()
```
